```python
import re
```

```python
from Bio import SeqIO
```

```python
fasta_sequence = SeqIO.parse("./chrM.fa",'fasta')
```

```python
for record in SeqIO.parse("./chrM.fa", "fasta"):
  print("%s %i" % (record.id, len(record)))
  name, sequence = record.id, str(record.seq)
pattern1="(?:G{3,}[ATGC]{1,7}){3,}G{3,}" 
pattern2="(?:C{3,}[ATGC]{1,7}){3,}C{3,}"
PQS_plus=[[m.start(),m.end(),m.group(0)] for m in re.finditer(pattern1,sequence.upper())]
PQS_minus=[[m.start(),m.end(),m.group(0)] for m in re.finditer(pattern2,sequence.upper())]
print(len(PQS_plus), len(PQS_minus))
```

Число найденных квадруплексов на + стренде равно 0, а на — стренде 8.

Теперь скачаем аннотацию генома и определим, в какие участки попадают Z-ДНК.

```python
!wget https://hgdownload.soe.ucsc.edu/goldenPath/ponAbe2/bigZips/genes/ponAbe2.refGene.gtf.gz
!gunzip ./ponAbe2.refGene.gtf.gz
```

```python
data2 = pd.read_csv("ponAbe2.refGene.gtf", sep = "\t", header = None)
```

```python
b_filtered_max = max(filtered_data[1])

def interval_ralation(a_filtered, b_filtered, a_data, b_data):    
    if a_filtered > b_filtered or a_data > b_data:
        print(a_filtered, b_filtered, a_data, b_data)
        return -1
    
    # filtered входит в data
    if a_filtered >= a_data and b_filtered <= b_data:
        return 2
    
    # не пересекаются 
    if (b_filtered <= a_data) or (a_filtered >= b_data):
        return 0
     
    # пересекаются 
    return 1

included = []
intersect = []

for d_idx in data2.index:
    
    d_a = data2[3][d_idx]
    d_b = data2[4][d_idx]
    
    if d_a > b_filtered_max:
        continue
        
    for f_idx in filtered_data.index:
        f_a = filtered_data[0][f_idx]
        f_b = filtered_data[1][f_idx]
        
        rel = interval_ralation(f_a, f_b, d_a, d_b)
        if rel == 2:
            included.append((f_idx, d_idx))
        elif rel == 1:
            intersect.append((f_idx, d_idx))

print(len(included))
print(len(intersect))
```

15 Z-ДНК попадают в гены, что составляет 50% от всех найденных Z-ДНК, 0 на пересечнии межгенного пространства и генов, остальные, значит, в межгенное пространство.

```python
included = []
intersect = []
b_pqs_max = 16482

for d_idx in data2.index:
    
    d_a = data2[3][d_idx]
    d_b = data2[4][d_idx]
    
    if d_a > b_pqs_max:
        continue
        
    for pqs in PQS_minus:
        pqs_a = pqs[0]
        pqs_b = pqs[1]
        
        rel = interval_ralation(f_a, f_b, d_a, d_b)
        if rel == 2:
            included.append((pqs_a, pqs_b, d_idx))
        elif rel == 1:
            intersect.append((pqs_a, pqs_b, d_idx))

print(len(included))
print(len(intersect))
```

```python
b_filtered_max = max(filtered_data[1])

def interval_ralation(a_filtered, b_filtered, a_data, b_data):    
    if a_filtered > b_filtered or a_data > b_data:
        print(a_filtered, b_filtered, a_data, b_data)
        return -1
    
    # filtered входит в data
    if a_filtered >= a_data and b_filtered <= b_data:
        return 2
    
    # не пересекаются 
    if (b_filtered <= a_data) or (a_filtered >= b_data):
        return 0
     
    # пересекаются 
    return 1

included_filt = []
intersect_filt = []

for d_idx in data3.index:
    
    d_a = data3[3][d_idx]
    d_b = data3[4][d_idx]
    
    if d_a > b_filtered_max:
        continue
        
    for f_idx in filtered_data.index:
        f_a = filtered_data[0][f_idx]
        f_b = filtered_data[1][f_idx]
        
        rel = interval_ralation(f_a, f_b, d_a, d_b)
        if rel == 2:
            included_filt.append((f_idx, d_idx))
        elif rel == 1:
            intersect_filt.append((f_idx, d_idx))

incl_dict = {}
for item in included_filt:
  if item[0] in incl_dict:
    incl_dict[item[0]] = incl_dict[item[0]] + 1
  else:
    incl_dict[item[0]] = 1

inters_dict = {}
for item in intersect_filt:
  if item[0] in inters_dict:
    inters_dict[item[0]] = inters_dict[item[0]] + 1
  else:
    inters_dict[item[0]] = 1

incl_count = len(incl_dict.keys())
inters_count = len(inters_dict.keys())

print(incl_dict.keys())
print(inters_dict.keys())
others = set(incl_dict.keys()) | set(inters_dict.keys())
print(others)

print(f"Z-ДНК попало в гены: {incl_count}, Z-ДНК попало в пересечения: {inters_count}, Z-ДНК попало в межгенное пространство: {len(filtered_data) - len(others)}")

print(len(included_filt))
print(len(intersect_filt))
```
